```python
# Comparsion and logical Ufuncs
a=np.array([1,2,3,4])
b=np.array([2,2,2,2])
print(np.less(a,b))
print(np.greater_equal(a,b))
print(np.equal(a,b))
print(np.not_equal(a,b))
```

```python
cond1=a>1
cond2=a<4
print(cond1)
print(cond2)
print(np.logical_and(cond1,cond2))
```

```python
print(np.median(stats_arr))
print(np.quantile(stats_arr,0.25))
```

```python
print(np.sum(stats_arr,axis=1))
print(np.sum(stats_arr,axis=0))
```
